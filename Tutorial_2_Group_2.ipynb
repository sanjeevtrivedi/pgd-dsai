## Chebyshev's Inequality

## 1.
A group of working professionals are surveyed and it is found that the average salary is 27LPA. It is also found that the standard deviation of salaries is 3 LPA.

- (a) Find the interval such that 99% of the salaries of the group lies in that range.
- (b) Is the graph of salaries symmetric about mean? Give reason.

### Solution:
We will solve this problem using Chebyshev's inequality.<br>
It is given that $$1-\frac{1}{k^2}\ge0.99$$
$$\Rightarrow \frac{1}{k^2}\le 0.01$$
$$\Rightarrow k^2\ge 100$$
$$\Rightarrow k>=10$$
because $k$ is positive.

It is given that $$|X-27|<3k=3\times 10=30$$.
Thus, the salaries lies in $[0,57)$ interval.

## 2.
A company has analyzed the weekly sales volumes at its retail locations and found that the average sales volume is 150 units with a standard deviation of 20 units.

(a) Calculate the range within which at least 75% of the weekly sales volumes fall. \
(b) Discuss whether it is possible to determine if the sales volume data is normally distributed from this information alone.

### Chebyshev's Inequality Example

**Given:** $\mu = 150 $ units, $\sigma = 20$ units

Chebyshev's inequality states:
$$
1 - \frac{1}{k^2} \geq 0.75 \quad \Rightarrow \quad \frac{1}{k^2} \leq 0.25 \quad \Rightarrow \quad k^2 \geq 4 \quad \Rightarrow \quad k \geq 2
$$

Using $k = 2$, the interval is:
$$
150 \pm 2 \times 20 = [110, 190]
$$
Thus, at least 75% of the weekly sales volumes fall between 110 and 190 units.

## 3.
An equilateral triangle has three zebras seated on each corner. Each zebra chooses a direction at random and only sprints along the triangle’s outline to either of the triangle’s opposing edges. Is there a chance that none of
the zebras will collide?

Given that all of the zebras are arranged in an equilateral triangle.

Note that if they’re sprinting down the outline to either edge, each zebra has two choices i.e. either a clockwise or counter-clockwise direction.

Let $Z_1^C, Z_1^N, Z_2^C, Z_2^N, Z_3^C, Z_3^N$ denote the zebra 1,2,3 clockwise and anticlockwise motion respectively.

Note that
$$P(Z_1^C)=P(Z_1^N) = P(Z_2^C) = P(Z_2^N) = P(Z_3^C) = P(Z_3^N) = \frac{1}{2}$$

$$\begin{align*}
P((Z_1^C \text{ and }Z_2^C \text{ and } Z_3^C) \text{ or } (Z_1^N \text{ and } Z_2^N \text{ and } Z_3^N)) &= P((Z_1^C \cap Z_2^C \cap Z_3^C)\cup (Z_1^N \cap Z_2^N \cap Z_3^N))\\
&= P(Z_1^C \cap Z_2^C \cap Z_3^C)+ P(Z_1^N \cap Z_2^N \cap Z_3^N)\\
&= P(Z_1^C)P(Z_2^C)P(Z_3^C)+ P(Z_1^N)P(Z_2^N)P(Z_3^N)\\
&= \frac{1}{2}\times \frac{1}{2}\times \frac{1}{2} + \frac{1}{2}\times \frac{1}{2}\times \frac{1}{2}\\
&= \frac{1}{4}
\end{align*}$$

## Empirical Rule

## 4.
You are given a dataset of annual rainfall in millimeters for a city over the last 20 years. The data is stored in a Python list:

```
# rainfall_data = [750, 800, 845, 890, 780, 920, 850, 880, 760, 840, 900, 830, 870, 910, 870, 950, 940, 850, 880, 790]

```

(a) Write a Python script to calculate the mean and standard deviation of the rainfall data. \
(b) Use the empirical rule to determine the range within which approximately 95% of the rainfall data should fall. \
(c) Plot the data using a histogram and mark the mean and the mean ± 2 standard deviations.

````python
import numpy as np
import matplotlib.pyplot as plt

# Given data
rainfall_data = [750, 800, 845, 890, 780, 920, 850, 880, 760, 840, 900, 830, 870, 910, 870, 950, 940, 850, 880, 790]
````

````python
# (a)
# Calculate the mean and standard deviation
mean_rainfall = np.mean(rainfall_data)
std_dev_rainfall = np.std(rainfall_data)

print(f"Mean Rainfall: {mean_rainfall:.2f} mm")
print(f"Standard Deviation: {std_dev_rainfall:.2f} mm")
````

````python
# (b)
# Empirical Rule to determine the range for 95% of the data
lower_bound = mean_rainfall - 2 * std_dev_rainfall
upper_bound = mean_rainfall + 2 * std_dev_rainfall

print(f"95% of the data falls between {lower_bound:.2f} mm and {upper_bound:.2f} mm")
````

````python
# (c)
# Plotting the data
plt.figure(figsize=(10, 6))
plt.hist(rainfall_data, bins=10, color='skyblue', alpha=0.7, label='Annual Rainfall')
plt.axvline(mean_rainfall, color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(lower_bound, color='green', linestyle='dashed', linewidth=2, label='Mean - 2SD')
plt.axvline(upper_bound, color='green', linestyle='dashed', linewidth=2, label='Mean + 2SD')
plt.title('Histogram of Annual Rainfall')
plt.xlabel('Rainfall (mm)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
````

## 5.
You have collected daily average temperature data (in degrees Celsius) for a city over the last month. The temperatures are as follows:


```
temperature_data = [22, 24, 23, 21, 22, 26, 27, 25, 23, 24, 25, 26, 24, 23, 22, 21, 22, 23, 25, 26, 27, 28, 27, 25, 24, 23, 22, 21, 22, 23]

```

(a) Write a Python script to calculate the mean and standard deviation of the temperature data. \
(b) Use the empirical rule to determine the range within which approximately 68% of the temperature data should fall. \
(c) Plot the data using a line graph and highlight the mean and the mean ± 1 standard deviation.

````python
import numpy as np
import matplotlib.pyplot as plt

# Given data
temperature_data = [22, 24, 23, 21, 22, 26, 27, 25, 23, 24, 25, 26, 24, 23, 22, 21, 22, 23, 25, 26, 27, 28, 27, 25, 24, 23, 22, 21, 22, 23]
````

````python
# (a)
# Calculate the mean and standard deviation
mean_temperature = np.mean(temperature_data)
std_dev_temperature = np.std(temperature_data)

print(f"Mean Temperature: {mean_temperature:.2f}°C")
print(f"Standard Deviation: {std_dev_temperature:.2f}°C")
````

````python
# (b)
# Empirical Rule to determine the range for 68% of the data
lower_bound = mean_temperature - std_dev_temperature
upper_bound = mean_temperature + std_dev_temperature

print(f"68% of the data falls between {lower_bound:.2f}°C and {upper_bound:.2f}°C")
````

````python
# (c)
# Plotting the data
plt.figure(figsize=(12, 6))
plt.plot(temperature_data, marker='o', color='blue', label='Daily Temperature')
plt.axhline(mean_temperature, color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.axhline(lower_bound, color='green', linestyle='dashed', linewidth=2, label='Mean - 1SD')
plt.axhline(upper_bound, color='green', linestyle='dashed', linewidth=2, label='Mean + 1SD')
plt.title('Daily Average Temperature Over a Month')
plt.xlabel('Day of the Month')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid(True)
plt.show()
````

## 6.
You are given the following dataset representing the **monthly study hours** and **exam scores** of 10 students.

### Dataset:

| Student | Study Hours | Exam Score |
|---------|------------|------------|
| A       | 2          | 50         |
| B       | 5          | 55         |
| C       | 7          | 60         |
| D       | 10         | 70         |
| E       | 12         | 72         |
| F       | 15         | 78         |
| G       | 18         | 85         |
| H       | 20         | 88         |
| I       | 22         | 92         |
| J       | 25         | 95         |

### Tasks:
- (a) Calculate the correlation coefficient between study hours and exam scores.
- (b) Plot a scatter plot of the dataset with a trendline.

````python
import numpy as np
import matplotlib.pyplot as plt

# Data
study_hours = np.array([2, 5, 7, 10, 12, 15, 18, 20, 22, 25])
exam_scores = np.array([50, 55, 60, 70, 72, 78, 85, 88, 92, 95])
````

````python
# (a)
# Compute correlation
correlation_matrix = np.corrcoef(study_hours, exam_scores)
correlation = correlation_matrix[0, 1]

correlation
````

````python
# (b)
# Plot the data
plt.figure(figsize=(8, 5))
plt.scatter(study_hours, exam_scores, color='blue', label='Data Points')
plt.xlabel('Study Hours')
plt.ylabel('Exam Scores')
plt.title(f'Study Hours vs Exam Scores (Correlation: {correlation:.2f})')
plt.grid(True)

# Add trendline
m, b = np.polyfit(study_hours, exam_scores, 1)
plt.plot(study_hours, m * study_hours + b, color='red', linestyle='--', label='Trendline')

plt.legend()
plt.show()
````

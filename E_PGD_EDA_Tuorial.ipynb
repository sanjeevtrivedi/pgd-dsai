# Handling Missing Values in Data

Missing values can be handled using different techniques depending on the nature of the data and the extent of missingness. Below are the most common approaches:

## 1. Drop Missing Values (`dropna()`)
- **What it does:** Removes rows (or columns) containing missing values.
- **When to use:**
  - If missing values are **very few** (e.g., <5% of the dataset).
  - If missing values are **randomly distributed** and dont introduce bias.
  - If the dataset is **large** and dropping rows doesnt significantly impact learning.
  - If the missing values occur in **non-critical columns**.
- **When NOT to use:**  
  - If a large portion of data is missing, as it may lead to data loss and poor model performance.

---

## 2. Mean Imputation (`fillna(df.mean())`)
- **What it does:** Replaces missing values with the **mean** of the respective column.
- **When to use:**
  - If the data is **normally distributed** (symmetric, no extreme outliers).
  - If missing values are **randomly distributed** and not correlated with the target variable.
- **When NOT to use:**  
  - If the data is **skewed** (e.g., salaries, house prices), as the mean is affected by outliers.

---

## 3. Median Imputation (`fillna(df.median())`)
- **What it does:** Replaces missing values with the **median** (middle value).
- **When to use:**
  - If the data contains **outliers** or is **skewed** (e.g., income, house prices).
  - If the data is **not normally distributed** and has a long tail.
- **When NOT to use:**  
  - If the data is symmetric and lacks extreme values, the **mean** might be a better choice.

---

## 4. Mode Imputation (`fillna(df.mode()[0])`)
- **What it does:** Replaces missing values with the **most frequent value**.
- **When to use:**
  - For **categorical data** (e.g., filling missing gender values with "Male" or "Female").
  - If certain values **dominate** the dataset, making the mode a good representative.
  - For **ordinal data** (e.g., education levels: High School, Bachelor's, Master's).
- **When NOT to use:**  
  - If the mode appears **very few times**, it may not represent missing values well.

---

## Choosing the Right Method
| Situation | Best Method |
|-----------|------------|
| Small % of missing values | Drop rows (`dropna()`) |
| Large % of missing values | Impute (mean, median, mode) |
| Numeric data (symmetric) | Mean |
| Numeric data (skewed) | Median |
| Categorical data | Mode |
| Critical missing values (e.g., target column) | Predictive Imputation (KNN, Regression) |

If your dataset has a structured pattern, **predictive imputation (e.g., KNN, regression)** may be more effective than simple imputation.

```python
Q1 = df.select_dtypes(include=np.number).quantile(0.25) # Select only numerical columns
Q3 = df.select_dtypes(include=np.number).quantile(0.75) # Select only numerical columns
IQR = Q3 - Q1
print(IQR)
```

```python
# Instead of using separate quantiles for all numeric columns,
# Calculate the quantiles and IQR for each numeric column individually within the DataFrame
numeric_cols = df.select_dtypes(include=np.number).columns

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    df = df[~((df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR)))]

df.shape
```

```python
from IPython.display import display, Math, Markdown

# Section 1: Min-Max Scaling
display(Markdown("# Normalization Techniques: Min-Max, Standardization, and Robust Scaling"))
display(Markdown("## 1. Min-Max Scaling (Feature Scaling)"))
display(Markdown("Min-Max scaling transforms data into a fixed range, typically $[0,1]$ or $[-1,1]$, using:"))
display(Math(r'X_{\text{scaled}} = \frac{X - X_{\min}}{X_{\max} - X_{\min}}'))
display(Markdown("- $X_{\text{scaled}}$ is the normalized value."))
display(Markdown("- $X$ is the original value."))
display(Markdown("- $X_{\min}$ and $X_{\max}$ are the minimum and maximum values of the feature."))
display(Markdown("### **When to Use Min-Max Scaling**"))
display(Markdown("- **Data has a defined range** (e.g., images, stock prices)."))
display(Markdown("- **Feature values should be bounded** (e.g., neural networks with sigmoid/tanh activation)."))
display(Markdown("- **No significant outliers** (since Min-Max is sensitive to extreme values)."))
display(Markdown("### **Case Scenarios**"))
display(Markdown("- **Image Processing:** Normalizing pixel values to [0,1] for stable training."))
display(Markdown("- **K-Means Clustering:** Ensures equal feature importance in distance calculations."))
display(Markdown("- **Neural Networks:** Prevents large weight updates when using activations like sigmoid or tanh."))

# Section 2: Standardization
display(Markdown("---"))
display(Markdown("## 2. Standardization (Z-score Normalization)"))
display(Markdown("Standardization scales data to have zero mean and unit variance:"))
display(Math(r'X_{\text{standardized}} = \frac{X - \mu}{\sigma}'))
display(Markdown("- $\mu$ is the mean."))
display(Markdown("- $\sigma$ is the standard deviation."))
display(Markdown("### **When to Use Standardization**"))
display(Markdown("- **Data is approximately normal (Gaussian distribution).**"))
display(Markdown("- **Features have different scales** (e.g., height in cm vs. weight in kg)."))
display(Markdown("- **Outliers exist** (since standardization is less affected than Min-Max)."))
display(Markdown("### **Case Scenarios**"))
display(Markdown("- **Principal Component Analysis (PCA):** Ensures equal variance contribution."))
display(Markdown("- **Linear Models (Logistic Regression, SVMs):** Assumes normally distributed residuals."))
display(Markdown("- **K-Nearest Neighbors (KNN):** Avoids biasing distance calculations toward larger magnitude features."))

# Section 3: Robust Scaling
display(Markdown("---"))
display(Markdown("## 3. Robust Scaling"))
display(Markdown("Robust Scaler transforms data based on the median and interquartile range (IQR), using:"))
display(Math(r'X_{\text{robust}} = \frac{X - Q_2}{Q_3 - Q_1}'))
display(Markdown("- $Q_2$ = Median"))
display(Markdown("- $Q_1, Q_3$ = First and third quartiles (25th and 75th percentiles)"))
display(Markdown("This method **centers data around the median** and **scales it based on IQR**, making it **robust to outliers**."))
display(Markdown("### **When to Use Robust Scaling**"))
display(Markdown("- **Data contains extreme outliers** that could distort Min-Max and Standardization."))
display(Markdown("- **Feature distributions are highly skewed** (e.g., salary distributions)."))
display(Markdown("- **Boxplot analysis shows heavy-tailed distributions.**"))
display(Markdown("### **Case Scenarios**"))
display(Markdown("- **Financial Data (Income, House Prices):** Helps when values have long-tail distributions."))
display(Markdown("- **Medical Data (Blood Pressure, Cholesterol Levels):** Accounts for natural outliers."))
display(Markdown("- **Logistic Regression on Skewed Data:** Ensures stable coefficients by reducing the impact of extreme values."))
```

```python
# Selecting numerical features for normalization
numerical_features = ["HP", "MPG-H", "Price"]

# Min-Max Normalization
min_max_scaler = MinMaxScaler()
df_min_max = df.copy()
df_min_max[numerical_features] = min_max_scaler.fit_transform(df[numerical_features])

# Standardization (Z-score normalization)
standard_scaler = StandardScaler()
df_standard = df.copy()
df_standard[numerical_features] = standard_scaler.fit_transform(df[numerical_features])

# Robust Scaling (Median & IQR-based scaling)
robust_scaler = RobustScaler()
df_robust = df.copy()
df_robust[numerical_features] = robust_scaler.fit_transform(df[numerical_features])

# Displaying the transformed data
print("Original Data:")
print(df[numerical_features])

print("\nMin-Max Scaled Data:")
print(df_min_max[numerical_features])

print("\nStandardized Data:")
print(df_standard[numerical_features])

print("\nRobust Scaled Data:")
print(df_robust[numerical_features])
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_squared_error,mean_absolute_error
```

```python
model1 = LinearRegression()
model2 = Ridge()
model3 = Lasso()
model4 = DecisionTreeRegressor()
```

```python
# train_df = df[numerical_features]
# train_df = df_min_max[numerical_features]
# train_df = df_standard[numerical_features]
train_df = df_robust[numerical_features]
from sklearn.model_selection import train_test_split
x = train_df.drop('Price', axis = 1)
y = train_df['Price']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)
accuracies=[]
def Accuracy_score(model):
    model.fit(x_train,y_train)
    predictions=model.predict(x_test)
    mae = mean_absolute_error(y_test, predictions)
    accuracies.append(mae)
    print(f'Mean Absolute Error: {mae}')

    mse = mean_squared_error(y_test, predictions)
    print(f'Mean Squared Error: {mse}')

    r2 = r2_score(y_test, predictions)
    print(f'R Score: {r2} \n')
```

You can see the result of changing the train_df , the robust scaler effectively tackles outliers

```python
Accuracy_score(model1) # LinearRegression
Accuracy_score(model2) # Ridge
Accuracy_score(model3) # Lasso
Accuracy_score(model4) # DecisionTree
```
